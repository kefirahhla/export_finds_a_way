```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import pandas as pd
import matplotlib.ticker as mtick
```

```python
X_to_KGZ2=pd.read_csv('/TradeData_6_7_2024_15_8_51.csv', sep=',', encoding='latin1') #2022
X_to_RUS2=pd.read_csv('/TradeData_6_7_2024_15_11_24.csv', sep=',', encoding='latin1')
X_to_KAZ2=pd.read_csv('/TradeData_6_7_2024_15_10_14.csv', sep=',', encoding='latin1')

X_to_KGZ2['Period']=X_to_KGZ2['Period']-202200
X_to_RUS2['Period']=X_to_RUS2['Period']-202200
X_to_KAZ2['Period']=X_to_KAZ2['Period']-202200

X_to_KGZ3=pd.read_csv('/TradeData_6_7_2024_15_12_50.csv', sep=',', encoding='latin1') #2023
X_to_RUS3=pd.read_csv('/TradeData_6_7_2024_15_12_6.csv', sep=',', encoding='latin1')
X_to_KAZ3=pd.read_csv('/TradeData_6_7_2024_15_13_33.csv', sep=',', encoding='latin1')

X_to_KGZ3['Period']=X_to_KGZ3['Period']-202300+12
X_to_RUS3['Period']=X_to_RUS3['Period']-202300+12
X_to_KAZ3['Period']=X_to_KAZ3['Period']-202300+12

X_to_KGZ = pd.concat([X_to_KGZ2, X_to_KGZ3], axis=0, ignore_index=True)
X_to_RUS = pd.concat([X_to_RUS2, X_to_RUS3], axis=0, ignore_index=True)
X_to_KAZ = pd.concat([X_to_KAZ2, X_to_KAZ3], axis=0, ignore_index=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
avg= XPgroup_to_KGZ.get_group('Belgium')['PrimaryValue'].mean()
Y_total = XPgroup_to_KGZ.get_group('Belgium')['PrimaryValue']/avg
X_total = pd.DataFrame(XPgroup_to_KGZ.get_group('Belgium'), columns=['Period'])
x_train, x_test, y_train, y_test = train_test_split(X_total, Y_total, test_size=0.2)
#.plot(kind='line', title='Primary Value', ax=axs[1])
LR = LinearRegression().fit(x_train, y_train)
LR.predict(x_test)
y_fin = LR.coef_* x_test + LR.intercept_
plt.scatter(x_test, y_fin, color="green")
plt.scatter(x_test, y_test, color="red")
plt.scatter(x_train, y_train, color="orange")
plt.show()
```

```python
fig, axs = plt.subplots(nrows=len(XPgroup_to_KGZ.groups)//3+1, ncols=3, figsize=(20, 70))
figsize=(20, 100)
d=XPgroup_to_KGZ.groups
dd=list(d.keys())
for i in range(len(XPgroup_to_KGZ.groups)):
  avg= XPgroup_to_KGZ.get_group(dd[i])['PrimaryValue'].mean()
  Y_total = XPgroup_to_KGZ.get_group(dd[i])['PrimaryValue']/avg
  X_total = pd.DataFrame(XPgroup_to_KGZ.get_group(dd[i]), columns=['Period'])
  x_train, x_test, y_train, y_test = train_test_split(X_total, Y_total, test_size=0.2, random_state=340)
  LR = LinearRegression().fit(x_train, y_train)
  LR.predict(x_test)
  x=range(1, 25)
  y_fin = LR.coef_* x + LR.intercept_
  axs[i//3, i%3].plot(x, y_fin, color="green")
  axs[i//3, i%3].scatter(x_test, y_test, color="red")
  axs[i//3, i%3].scatter(x_train, y_train, color="orange")
  axs[i//3, i%3].set_xlabel('Period')
  axs[i//3, i%3].set_ylabel('Scaled export')
  axs[i//3, i%3].set_title(dd[i])

plt.show()
```
